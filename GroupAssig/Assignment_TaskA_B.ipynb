```python
# Import libraries
import pandas as pd
import pymongo
from pymongo import MongoClient
import csv, json
from pprint import pprint
```

```python
# Read CSV into dataframes
climate_df = pd.read_csv("climate_historic.csv")
hotspot_df = pd.read_csv("hotspot_historic.csv")

# Create a function for converting file type (CSV to JSON)
def convert_csv_to_json(csv_df, json_file):
    json_export = csv_df.to_json(json_file, orient='records')
    with open(json_file) as file:
        data = json.load(file)
    
    return data
```

```python
# Get the data
climate_data = convert_csv_to_json(climate_df, "climate_historic.json") # Get the climate data
hotspot_data = convert_csv_to_json(hotspot_df, "hotspot_historic.json") # Get the hotspot data
    
# Connect to MongoDB
client = MongoClient () # Connect on the default host and port
db = client.fit5148_db # Create a new database or get the existed database (for creating, you have to insert data first and then you will see the new database)

# Create collections
climate_collection = db.climate # Create climate collection
hotspot_collection = db.hotspot # Create hotspot collection

# Insert date into collections
climate_result = climate_collection.insert_many(climate_data) # Insert data into climate collection
hotspot_result = hotspot_collection.insert_many(hotspot_data) # Insert data into hotspot collection
```

```python
# Create a function for checking if the data exists in the collection now
def check_collection(collection_name): 
    cursor = collection_name.find({})
    for document in cursor: 
        pprint(document)
```

```python
# Check climate collection
check_collection(climate_collection)
```

```python
# Check hotspot collection
check_collection(hotspot_collection)
```

## Task B. Querying MongoDB using PyMongo (30 Marks)
1. Find climate data on 10th December 2017.
2. Find the latitude, longitude, surface temperature (C), and confidence when the surface temperature (C) was between 65 C and 100 C.
3. Find date, surface temperature (C), air temperature (C), relative humidity and max wind speed on 15th and 16th of December 2017.
4. Find datetime, air temperature (C), surface temperature (C) and confidence when the confidence is between 80 and 100.
5. Find the top 10 records with the highest surface temperature (C).
6. Find the number of fire in each day. You are required to only display the total number of fire and the date in the output.
7. Find the average surface temperature (C) for each day. You are required to only display average surface temperature (C) and the date in the output.

```python
# Show climate collection
allDate = climate_collection.find()
for record in allDate:
    pprint(record)
```

#### 1. Find climate data on 10th December 2017.

```python
results = climate_collection.find({"date":"10/12/2017"})
for doc in results:
    pprint(doc)
```

#### 2. Find the latitude, longitude, surface temperature (C), and confidence when the surface temperature (C) was between 65 C and 100 C.

```python
results = hotspot_collection.find({"$and":[{"surface_temperature_celcius":{"$gte":65}}, {"surface_temperature_celcius":{"$lte":100}}]})
for doc in results:
    pprint(doc)
```

#### 3. Find date, surface temperature (C), air temperature (C), relative humidity and max wind speed on 15th and 16th of December 2017.

```python
# Join 2 dataframe based on the common date
joined_df = pd.merge(climate_df, hotspot_df, on='date', how='outer')

# Get the joined data
joined_data = convert_csv_to_json(joined_df, "joined_data.json")

# Create joined collection
joined_collection = db.joined

# Insert data into joined collection
joined_result = joined_collection.insert_many(joined_data)

# Check joined collection
check_collection(joined_collection)
```

```python
results = joined_collection.find({"$or":[{"date":"15/12/2017"}, {"date":"16/12/2017"}]}, {"date":1, "surface_temperature_celcius":1, "air_temperature_celcius":1, "relative_humidity":1, "max_wind_speed":1})
for doc in results:
    pprint(doc)
```

#### 4. Find datetime, air temperature (C), surface temperature (C) and confidence when the confidence is between 80 and 100.

```python
results = joined_collection.find({"$and":[{"confidence":{"$gte":80}}, {"confidence":{"$lte":100}}]}, {"air_temperature_celcius":1, "surface_temperature_celcius":1, "confidence":1})
for doc in results:
    pprint(doc)
```

#### 5. Find the top 10 records with the highest surface temperature (C).

```python
results = hotspot_collection.find().sort("surface_temperature_celcius",-1).limit(10)
for doc in results:
    pprint(doc)
```

#### 6. Find the number of fire in each day. You are required to only display the total number of fire and the date in the output.

```python
pipeline = [
    { "$group": { "_id": "$date", "count": { "$sum" : 1 } } }
]

for doc in hotspot_collection.aggregate(pipeline):
    print(doc)
```

#### 7. Find the average surface temperature (C) for each day. You are required to only display average surface temperature (C) and the date in the output.

```python
pipeline = [
    { "$group": { "_id": "$date", "Average temp": { "$avg": "$surface_temperature_celcius" } } }
]

for doc in hotspot_collection.aggregate(pipeline):
    print(doc)
```
